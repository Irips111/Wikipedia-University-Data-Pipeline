```python
table = soup.find('table', attrs={'class': 'sortable'})
```

```python
trs = table.find_all('tr')
```

```python
trs[0]
```

```python
columns = []
for item in trs[0].find_all('th'):
    print(item.text.strip())
    columns.append(item.text.strip())
```

```python
columns
```

```python
trs[0].find_all('th')
```

```python
columns = list(map(lambda x: x.text.strip(), trs[0].find_all('th')))
columns[-1] = "Link"
```

```python
rows = trs[1:]
```

```python
rows[0]
```

```python
rows[0].find_all('td')
```

```python
list(map(lambda x: x.text.strip(), rows[0].find_all('td')))
```

```python
rows[0].find_all('td')[1].a.attrs['href'].lstrip('/')
```

```python
def extract_row(tr):
    row_soup_list = tr.find_all('td')
    row = list(map(lambda x: x.text.strip(), row_soup_list))
    if len(row) < 6:
        row.append('')
        print("===========================================================================")
        print(row)
    link = row_soup_list[1].a.attrs['href'].lstrip('/')
    row[-1] = f'https://en.wikipedia.org/{link}' if link else ''
    return row
```

```python
extract_row(rows[35])
```

```python
data = list(map(extract_row, rows))
```

```python
import csv
import json
import pandas as pd
```

```python
with open('universities.csv', 'w', newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(columns)
    csvwriter.writerows(data)
```

```python
with open('universities.json', 'w') as f:
    json.dump(data, f, indent=4)
```
